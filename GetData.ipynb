```python
import pandas as pd
import datetime
import time
from urllib.error import HTTPError
```

```python
def GetPricesInfo(date: datetime.date, entries: int) -> pd.DataFrame:
    url = 'https://coinmarketcap.com/historical/'
    url = url + date.strftime('%Y%m%d')
    data = pd.read_html(url)[-1].head(entries)
    data = data[['Rank', 'Name', 'Symbol', 'Market Cap', 'Price']]
    data['Rank'] = data['Rank'].apply(lambda x: int(x))
    data['Market Cap'] = data['Market Cap'].apply(lambda x: float(x.replace('$', '').replace(',', '')))
    data['Price'] = data['Price'].apply(lambda x: float(x.replace('$', '').replace(',', '')))
    data['Date'] = date
    return data
```

```python
date = datetime.date(2017, 3, 19)
print('Gettting data for ' + str(date))
while date < datetime.date.today():
    try:
        df = GetPricesInfo(date, 20)
    except HTTPError as e:
        if e.code == 429:
            print('Too many request error, going to sleep for a minute...')
            time.sleep(60)
        else:
            raise
    else:
        df.to_csv('Data/' + str(date) + '.csv')
        date += datetime.timedelta(days=7)
        print('Gettting data for ' + str(date))
```
