Try writing an algorithm for re-balancing the index based on the coins momentums. 
For example, measure the growth of each coin over a period of time, and if a coin is performing poorly in relation to the other coins, we need to change the percentages.

```python
initial_balance = 1000
```

```python
index = Index('My Index', 'Liav')
```

```python
index.AddSymbol('BTC', 0.5)
index.AddSymbol('ETH', 0.3)
index.AddSymbol('SOL', 0.2)
```

```python
index.RemoveSymbol('ETH')
```

```python
index.UpdateWeight('SOL', 0.5)
```

```python
balance_progress = index.Backtest(initial_balance, 0.001)
```

```python
len(balance_progress[1])
```

```python
print(index.symbols_weights)
print(f'Profit percent: {((balance_progress[-1] - initial_balance)/initial_balance)*100}%')
fig = plt.figure(figsize=(20,8))
plt.plot(balance_progress)
```

```python
top_index = TopCryptoIndex(10, 0.75)
```

```python
balance_progress = top_index.Backtest(initial_balance, 0.001)
```

```python
print(f'Profit percent: {((balance_progress[-1] - initial_balance)/initial_balance)*100}%')
fig = plt.figure(figsize=(20,8))
plt.plot(balance_progress)
```
